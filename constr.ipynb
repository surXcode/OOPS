```python
class Employee:

    def __init__(self, eid, ename, ephonno, esal, eadd):
        self.eid = eid
        self.ename = ename
        self.ephonno = ephonno
        self.esal = esal
        self.eadd = eadd


e1 = Employee("E001", "Saurabh", 987654321, 200000, "Kanpur")
e2 = Employee("E002", "Suraj", 987654322, 2333000, "Sultanpur")
e3 = Employee("E003", "Anuj", 9876333321, 245600, "Noida")

emp_list = [e1, e2, e3]

for emp in emp_list:
    print(f"Employee Name: {emp.ename}")
    print(f"Employee Address: {emp.eadd}")
    print("=" * 30)
```
