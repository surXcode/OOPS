# Methods: function inside of a class is known as method.

# There are three types of methods:

1. class method
2. object method/instance method
3. static method

1. class method:- It is use to access and modify the class properties.

* classmethod decorator is use before  defining a class method.

* Symbol: #classmethod

* it take the cls first arguments

* cls carries the address of the class or object.

** Syntax: **
- class classname:
- @classmethod
- def func.name(cls)
- SB
- obj = classname()
- classname func_name()
- obj.func_name()


```python
class Bank:
    bname = "BOB"
    loc = "Noida"
    Ifsc = 'BOB123B'
    bmanager = 'Mr.Ashvik'

    @classmethod
    def banks_details(cls):
        print(cls.bname)
        print(cls.loc)
        print(cls.Ifsc)
        print(cls.bmanager)

    @classmethod
    def change_loc(cls):
        cls.loc = input("Enter the new location :")
        print("Location is change successfully :")
        cls.banks_details()




c1 = Bank()
c2 = Bank()
Bank.change_loc()

```

2. object method/instance method:

- It takes mandatory argument the shell.

- self carries the address of object.

- it is use to access and modify.

```python
class Bank:
    bname = "BOB"
    loc = "Noida"
    Ifsc = 'BOB123B'
    bmanager = 'Mr.Ashvik'

    def __init__(self, name, addr, sal):
        self.name = name
        self.addr = addr
        self.sal = sal

    @classmethod
    def banks_details(cls):
        print(f"Bank name     : {cls.bname}")
        print(f"Location      : {cls.loc}")
        print(f"IFSC code     : {cls.Ifsc}")
        print(f"Bank manager  : {cls.bmanager}")

    def change_address(self):
        self.addr = input("Enter the new address of the customer : ")
        print("Address is changed successfully.")



c1 = Bank('Ashvik', 'Mumbai', 100000)
c2 = Bank('Abhinandan', 'Noida', 200000)


Bank.banks_details()

c1.change_address()
print(c1.addr)

c1.change_address()
print("Updated Address:", c1.addr)
```

# Static Method:

- It is neither related to class nor instance (object).

- It is use to store extra functionality.

- It is used us utility or helper function inside the class. 

- it is does 'nt take any mandatory argument like - cls and 

- static method decorator is use to define the static method.

# Syntax:

- class class_name:
-    @staticmethod
-     def method_name():
-         SB


- class method_name()
- obj.method_name()

```python
class Student:
    
    
    def __init__(self, name, roll, marks):
        self.name = name          
        self.roll = roll          
        self.marks = marks        

    
    def display_info(self):
        print("Student Name :", self.name)
        print("Roll Number  :", self.roll)
        print("Marks        :", self.marks)

    
    def grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        elif self.marks >= 50:
            return "C"
        else:
            return "D"



s1 = Student("Ashvki", 101, 85)
s2 = Student("Rahul", 102, 45)


s1.display_info()
print("Grade:", s1.grade())

print()

s2.display_info()
print("Grade:", s2.grade())
```
