# There are five types of inheritance: 

1. Single level inheritance
2. Multiple level inheritance
3. Multilevel inheritance
4. Hierarchical Inheritance
5. Hybrid inheritance


1. Single level inheritance: 
 
- Here, properties are transferred from a single parent class to a single child class, but the reverse is not possible.

# Syntax:
- class Parent:
                ----------
                | SB      |
                 ---------
- class Child(Parent):
                ----------
                | SB     |
                 ---------







```python
class Main:
    a = 100
    b = "hello"


class Demo:
    a = 200
    b = "Bye"


class New(Demo):
    c = "This is child class"


ob1 = Main()
ob2 = Demo()
ob3 = New()

print(ob1.a)  
print(ob2.b)   
print(ob3.c)   
```

2. Multilevel Inheritance:

- In multilevel inheritance, one class inherits from another class, and then another class inherits from that class.

# Syntax: 
- class A:          # Parent Class
    pass

- class B(A):       # Child of A
    pass

- class C(B):       # Child of B (Grandchild of A)
    pass


# Diagram:
- A   →   Parent Class
 ↑
 |
- B   →   Child of A
 ↑
 |
- C   →   Child of B (Grandchild of A)

```python
class First:
    a = 10
    b = 20


class Second(First):
    x = "hii"
    y = 'Hello'


class Third(Second):
    j = True
    k = False


ob1 = Third()
print(ob1.a)
print(ob1.b)
print(ob1.x)
print(ob1.y)
print(ob1.j)
print(ob1.k)
```

# Multiple Inheritance:

# Syntax:

# Method resolution order: 

```python
class Addition:

    @staticmethod
    def add(a, b):
        return a + b


class Subtraction:

    @staticmethod
    def sub(a, b):
        return a - b


class Product:

    @staticmethod
    def pro(a, b):
        return a * b


class Calculator(Addition, Subtraction, Product):

    @staticmethod
    def div(a, b):
        return a / b


ob1 = Calculator()
print(ob1.add(10, 20))
print(ob1.sub(100, 20))
print(ob1.pro(10, 20))
print(ob1.div(10, 20))
```

# Hybrid Inheritance:- 

- It is the combination of two or more than two types of inheritance.

# Syntax:

       Class A
      /       \

```python
class upper:

    @staticmethod
    def uppercase():
        res = ''
        for i in range(ord('A'), ord('Z') + 1):
            res += chr(i)
        return res
class Alpha(upper):

    @staticmethod
    def lowercase():
        res = ''
        for i in range(ord('a'), ord('z') + 1):
            res += chr(i)
        return res
```

# Constructor chainnig:-

- It is phenomenon of calling the parent class constructor inside the child class.

- if we want to add extra properties for the child class onbject then we should use constructor chaining.

- We can do so by using super function or class name.

- constructor call should be done inside the constructor of the child class.



# Syntax:

                +-------------------------+
                |        CLASS            |
                |   -------------------   |
                |   |   Constructor   |   |
                |   |   __init__()    |   |
                |   -------------------   |
                |          |              |
                +----------|--------------+
                           |
                           | Object Creation
                           v
                +-------------------------+
                |        OBJECT           |
                |  (Memory Allocated)     |
                |   Properties Initialized |
                +-------------------------+



```python
class Demo:

    def __init__(self):
        print("This is demo class constructor")


class Primary(Demo):

    def __init__(self):
        super().__init__()
        print("This is primary class constructor")


ob1 = Primary()


# Constructor calling by using class
class A:

    def __init__(self):
        print("This is a constructor of A")


class B:

    def __init__(self):
        print("This is a constructor of B")


class C:

    def __init__(self):
        print("This is a constructor of C")


class Main(A, B, C):

    def __init__(self):
        C.__init__(self)
ob1 =Main()
```
